```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_file = open("cnn_eval_map.txt", "r")
fi = my_file.read()

nums = []
for i in fi.strip().split('\n'):
    nums.append(float(i))
    
nums_2 = []
my_file = open("cnn_train_map.txt", "r")
fi = my_file.read()

for i in fi.strip().split('\n'):
    nums_2.append(float(i))

fig = plt.figure(figsize = (15,10))
ax1 = fig.add_subplot()
plot_title = 'Mean Average Precision over epochs for CNN'
x_label_title = 'Epochs'
y_label_title = 'mAP'
plot_name = 'cnn_training_map'
plt.title(plot_title, fontsize=20)
ax1.plot(nums_2, label = 'Train mAP')
ax1.plot(nums, label = 'Eval mAP')
plt.legend()
ax1.set_xlabel(x_label_title, fontsize=20)
ax1.set_ylabel(y_label_title, fontsize=20)
plt.savefig(f'plots/{plot_name}.pdf')
```
